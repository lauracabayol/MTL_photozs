```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import os

from scipy import interpolate

from torch import nn, optim
from torch.optim import lr_scheduler

from scipy.integrate import simps

import time
```

```python
from astropy import units as u
from astropy.coordinates import SkyCoord
```

```python
def sigma68(data): return 0.5*(pd.Series(data).quantile(q = 0.84) - pd.Series(data).quantile(q = 0.16))
```

## DATA CATALOGUE

```python
query = f"""
    SELECT l.u_mag_aper2, l.b_mag_aper2, l.v_mag_aper2, l.r_mag_aper2, l.ip_mag_aper2, l.zp_mag_aper2, l.alpha_j2000, l.delta_j2000, l.photoz
    FROM cosmos2015_laigle_v1_1 as l"""

from sqlalchemy.engine import create_engine
from sqlalchemy import *
from sqlalchemy.orm import *

engine = create_engine('postgresql://readonly@db03.pau.pic.es/dm')


with engine.begin() as conn:
    conn.execute("SET TRANSACTION ISOLATION LEVEL SERIALIZABLE READ ONLY DEFERRABLE")
    laigle = pd.read_sql(query, conn)
```

```python
query = f"""
    SELECT cm.paudm_id, cm.zspec, cm."I_auto", cm.ra, cm.dec
    FROM COSMOS as cm
    """

with engine.begin() as conn:
    conn.execute("SET TRANSACTION ISOLATION LEVEL SERIALIZABLE READ ONLY DEFERRABLE")
    ilbert = pd.read_sql(query, conn)
```

```python
laigle_coords = SkyCoord(laigle.alpha_j2000.values*u.deg,laigle.delta_j2000.values*u.deg)
ilbert_coords = SkyCoord(ilbert.ra.values*u.deg,ilbert.dec.values*u.deg)

idx, sep2d, _ = laigle_coords.match_to_catalog_sky(ilbert_coords)
```

```python
laigle['idx'] = idx
laigle['sep'] = sep2d.arcsec
laigle['paudm_id'] = ilbert.loc[idx,'paudm_id'].values
laigle['iauto'] = ilbert.loc[idx,'I_auto'].values

laigle['Flag'] = np.zeros(shape = len(laigle))
laigle.Flag.where(laigle.sep > 1, True, inplace=True)

laigle = laigle[laigle.Flag == True]
```

```python
laigle = laigle.merge(ilbert, on = ['paudm_id'])
```

```python
laigle = laigle.replace(-99.9,np.nan)
laigle = laigle.dropna()
laigle = laigle.drop_duplicates('paudm_id')
laigle = laigle[laigle.zspec < 1.5]

catalog = laigle.copy()
catalog = catalog[catalog.iauto < 23]
```

```python
z = pd.read_csv('/data/astro/scratch/lcabayol/Euclid/cosmos_pz_cat_alex.csv', comment = '#', header = 0, sep = ',')
z = z.rename(columns = {'ref_id':'paudm_id'})
z = z.rename(columns = {'photoz':'zb_alex'})
z = z[z.zb_alex <1.5]
z = z.replace('N',np.nan, regex=True)
z = z.astype(np.float32)
```

```python
catalog = catalog.merge(z[['paudm_id','zb_alex']], on = ['paudm_id'])
```

```python
#zcfht = pd.read_csv('/cephfs/pic.es/astro/scratch/lcabayol/Euclid/data/photoz_cfht.txt')
zcfht = pd.read_csv('/data/astro/scratch/lcabayol/Euclid/Hildebrant_photozs.csv', header = 0, sep = ',')

cat_coords = SkyCoord(catalog.ra.values*u.deg,catalog.dec.values*u.deg)
cfht_coords = SkyCoord(zcfht.ra.values*u.deg,zcfht.dec.values*u.deg)

idx, sep2d, _ = cat_coords.match_to_catalog_sky(cfht_coords)

catalog['idx'] = idx
catalog['sep'] = sep2d.arcsec
catalog['cfht_pz'] = zcfht.loc[idx,'zb'].values

catalog['Flag'] = np.zeros(shape = len(catalog))
catalog.Flag.where(catalog.sep > 1, True, inplace=True)

catalog = catalog[catalog.Flag == True]
catalog = catalog[catalog.cfht_pz != 99]
```

```python
catalog_spec = catalog[catalog.zspec != 0].reset_index()
```

```python
dispersions = np.linspace(0,0.06,20)
```

```python
disp_data = []
rerr = (df_disp_data.loc[:,'zb'] - df_disp_data.zb_true) / (1+ df_disp_data.zb_true)
disp_data.append(sigma68(rerr))

for k in range(1,20):
    rerr = (df_disp_data.loc[:,'zb_%s'%k] - df_disp_data.zb_true) / (1+ df_disp_data.zb_true)
    disp_data.append(sigma68(rerr))
    
```

```python
disp_sims = []
rerr = (df_disp_sims.loc[:,'zb'] - df_disp_sims.zb_true) / (1+ df_disp_sims.zb_true)
disp_sims.append(sigma68(rerr))

for k in range(1,20):
    rerr = (df_disp_sims.loc[:,'zb_%s'%k] - df_disp_sims.zb_true) / (1+ df_disp_sims.zb_true)
    disp_sims.append(sigma68(rerr))
    
```
