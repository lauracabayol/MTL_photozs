```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import os

from torch import nn, optim
from torch.optim import lr_scheduler
import time
import SOM
```

```python
def sigma68(data): return 0.5*(pd.Series(data).quantile(q = 0.84) - pd.Series(data).quantile(q = 0.16))
```

```python
from sqlalchemy.engine import create_engine
from sqlalchemy import *
from sqlalchemy.orm import *

engine = create_engine('postgresql://readonly@db03.pau.pic.es/dm')

query = f"""
    SELECT cm.paudm_id, cm.zspec, cm."I_auto", cm.ra, cm.dec, cm.conf, cm."Umag", cm."Bmag", cm."Vmag", cm."Rmag", cm."ICmag", cm."Zmag"
    FROM COSMOS as cm
    """

with engine.begin() as conn:
    conn.execute("SET TRANSACTION ISOLATION LEVEL SERIALIZABLE READ ONLY DEFERRABLE")
    ilbert = pd.read_sql(query, conn)
```

```python
df_bb = pd.read_csv('/data/astro/scratch/lcabayol/Euclid/referee_report/catalogues/df_bb.csv', header = 0, sep =',')
df_bbnb = pd.read_csv('/data/astro/scratch/lcabayol/Euclid/referee_report/catalogues/df_bbnb.csv', header = 0, sep =',')
df_bbz = pd.read_csv('/data/astro/scratch/lcabayol/Euclid/referee_report/catalogues/df_bbz.csv', header = 0, sep =',')
df_bbnbz = pd.read_csv('/data/astro/scratch/lcabayol/Euclid/referee_report/catalogues/df_bbnbz.csv', header = 0, sep =',')
```

```python
df_bb = df_bb.rename(columns = {'zb':'zb_bb','rerr':'rerr_bb'})
df_bbnb = df_bbnb.rename(columns = {'zb':'zb_bbnb','rerr':'rerr_bbnb'})
df_bbz = df_bbz.rename(columns = {'zb':'zb_bbz','rerr':'rerr_bbz'})
df_bbnbz = df_bbnbz.rename(columns = {'zb':'zb_bbnbz','rerr':'rerr_bbnbz'})
```

```python
imag = ilbert[['paudm_id','I_auto','zspec']]
df_bb = df_bb.merge(imag, right_on = 'paudm_id', left_on = 'ref_id')
df = df_bb.merge(df_bbnb[['ref_id','zb_bbnb','rerr_bbnb']], on='ref_id')
df = df.merge(df_bbz[['ref_id','zb_bbz','rerr_bbz']], on='ref_id')
df = df.merge(df_bbnbz[['ref_id','zb_bbnbz','rerr_bbnbz']], on='ref_id')
```

```python
ilbert_sample = ilbert[ilbert.paudm_id.isin(df.ref_id)]
```

```python
ilbert_bb = ['Umag','Bmag','Vmag','Rmag','ICmag','Zmag']
```

## Initialise and train SOM

```python
nx,ny = 40,40
epochs = 100000
```

```python
selforgmap = SOM.SOM(x = nx, y = ny, epochs = epochs, vec_size = 5, metric = 'Euclidean',lr0 = 0.01, sigma0 = 5, initialisation = 'RandomNormal')
```

```python
training_sample = ilbert_sample[ilbert_bb].values[:,:-1] - ilbert_sample[ilbert_bb].values[:,1:]

selforgmap.train(training_sample, samples_epoch = 1)
```

```python
test_sample = ilbert_sample[ilbert_bb].values[:,:-1] - ilbert_sample[ilbert_bb].values[:,1:]
bmu = selforgmap.test_obj(test_sample)
```

```python
SOM_cosmos = pd.read_csv('/data/astro/scratch/lcabayol/Euclid/SOM_COSMOS.csv', sep = ',', header = 0)
SOM_ids = pd.read_csv('/data/astro/scratch/lcabayol/Euclid/SOMS_ids.csv', sep = ',', header = 0) 
SOM_cosmos = SOM_cosmos.rename(columns = {'ref_id':'paudm_id'})
```

```python
#ilbert_sample['bmu'] = bmu
ilbert_sample = ilbert_sample.merge(SOM_cosmos, on = 'paudm_id') 
nx, ny = 70, 60

x_cells = np.arange(0,nx)
y_cells = np.arange(0,ny)
index_cell = np.arange(nx*ny)
cells = np.array(np.meshgrid(x_cells,y_cells)).T.reshape(-1,2)
bmu_mapping = pd.DataFrame(np.c_[cells[:,0],cells[:,1],index_cell], columns = ['x_cell','y_cell','bmu'])

```

```python
ilbert_sample = ilbert_sample.merge(bmu_mapping, on = 'bmu')
```

```python
df = df.merge(ilbert_sample[['paudm_id','bmu','x_cell','y_cell']], left_on ='ref_id', right_on = 'paudm_id')
        

        
```

```python
som_zmean = df.groupby('bmu').zb_true.median().reset_index()#.rename(columns = {'zb_true':'mu_som'})
som_bbmean = df.groupby('bmu').zb_bb.median().reset_index()#.rename(columns = {'zb_true':'mu_som'})
som_bbnbmean = df.groupby('bmu').zb_bbnb.median().reset_index()#.rename(columns = {'zb_true':'mu_som'})
som_bbnbzmean = df.groupby('bmu').zb_bbnbz.median().reset_index()#.rename(columns = {'zb_true':'mu_som'})

som_bbbias = df.groupby('bmu').rerr_bb.median().reset_index()#.rename(columns = {'zb_true':'mu_som'})
som_bbnbbias = df.groupby('bmu').rerr_bbnb.median().reset_index()#.rename(columns = {'zb_true':'mu_som'})
som_bbnbzbias = df.groupby('bmu').rerr_bbnbz.median().reset_index()#.rename(columns = {'zb_true':'mu_som'})

som_bbscatter = df.groupby('bmu').rerr_bb.apply(sigma68).reset_index()#.rename(columns = {'zb_true':'mu_som'})
som_bbnbscatter = df.groupby('bmu').rerr_bbnb.apply(sigma68).reset_index()#.rename(columns = {'zb_true':'mu_som'})
som_bbnbzscatter = df.groupby('bmu').rerr_bbnbz.apply(sigma68).reset_index()#.rename(columns = {'zb_true':'mu_som'})


som_bbdispersion = df.groupby('bmu').zb_bb.apply(sigma68).reset_index()#.rename(columns = {'zb_true':'mu_som'})
som_bbnbdispersion = df.groupby('bmu').zb_bbnb.apply(sigma68).reset_index()#.rename(columns = {'zb_true':'mu_som'})
som_bbnbzdispersion = df.groupby('bmu').zb_bbnbz.apply(sigma68).reset_index()#.rename(columns = {'zb_true':'mu_som'})

```

```python
som_zmean = som_zmean.merge(bmu_mapping, on = 'bmu')
som_zmean = som_zmean.pivot(index = 'x_cell', columns = 'y_cell', values = 'zb_true')

som_bbmean = som_bbmean.merge(bmu_mapping, on = 'bmu')
som_bbmean = som_bbmean.pivot(index = 'x_cell', columns = 'y_cell', values = 'zb_bb')

som_bbnbmean = som_bbnbmean.merge(bmu_mapping, on = 'bmu')
som_bbnbmean = som_bbnbmean.pivot(index = 'x_cell', columns = 'y_cell', values = 'zb_bbnb')

som_bbnbzmean = som_bbnbzmean.merge(bmu_mapping, on = 'bmu')
som_bbnbzmean = som_bbnbzmean.pivot(index = 'x_cell', columns = 'y_cell', values = 'zb_bbnbz')
#----------------
som_bbbias = som_bbbias.merge(bmu_mapping, on = 'bmu')
som_bbbias = som_bbbias.pivot(index = 'x_cell', columns = 'y_cell', values = 'rerr_bb')

som_bbnbbias = som_bbnbbias.merge(bmu_mapping, on = 'bmu')
som_bbnbbias = som_bbnbbias.pivot(index = 'x_cell', columns = 'y_cell', values = 'rerr_bbnb')

som_bbnbzbias = som_bbnbzbias.merge(bmu_mapping, on = 'bmu')
som_bbnbzbias = som_bbnbzbias.pivot(index = 'x_cell', columns = 'y_cell', values = 'rerr_bbnbz')
#------------------
som_bbscatter = som_bbscatter.merge(bmu_mapping, on = 'bmu')
som_bbscatter = som_bbscatter.pivot(index = 'x_cell', columns = 'y_cell', values = 'rerr_bb')

som_bbnbscatter = som_bbnbscatter.merge(bmu_mapping, on = 'bmu')
som_bbnbscatter = som_bbnbscatter.pivot(index = 'x_cell', columns = 'y_cell', values = 'rerr_bbnb')

som_bbnbzscatter = som_bbnbzscatter.merge(bmu_mapping, on = 'bmu')
som_bbnbzscatter = som_bbnbzscatter.pivot(index = 'x_cell', columns = 'y_cell', values = 'rerr_bbnbz')
#-------------------
som_bbdispersion = som_bbdispersion.merge(bmu_mapping, on = 'bmu')
som_bbdispersion = som_bbdispersion.pivot(index = 'x_cell', columns = 'y_cell', values = 'zb_bb')

som_bbnbdispersion = som_bbnbdispersion.merge(bmu_mapping, on = 'bmu')
som_bbnbdispersion = som_bbnbdispersion.pivot(index = 'x_cell', columns = 'y_cell', values = 'zb_bbnb')

som_bbnbzdispersion = som_bbnbzdispersion.merge(bmu_mapping, on = 'bmu')
som_bbnbzdispersion = som_bbnbzdispersion.pivot(index = 'x_cell', columns = 'y_cell', values = 'zb_bbnbz')

```

## emission lines

```python
from astropy.cosmology import FlatLambdaCDM
import astropy.units as u
cosmo = FlatLambdaCDM(H0=70 * u.km / u.s / u.Mpc, Tcmb0=2.725 * u.K, Om0=0.3)
Ld = cosmo.luminosity_distance(som_zmean.values).T
```

```python
zalex_photz = pd.read_csv('/data/astro/scratch/lcabayol/Flux_estimation/data/cats_all/zalex_phot.csv', sep =',', header = 0, comment = '#')
zalex_photometry = pd.read_csv('/data/astro/scratch/lcabayol/Euclid/zalex_photometry.csv', sep =',', header = 0, comment = '#')
zalex_photometry= zalex_photometry[zalex_photometry.band =='i_Subaru'][['ref_id','flux','flux_err']]
```

```python
zalex_photz = zalex_photz.replace('N',np.nan, regex=True)
zalex_photz = zalex_photz.astype(np.float32)
```

```python
zalex_photz = zalex_photz.merge(zalex_photometry, on = 'ref_id')

zalex_photz = zalex_photz[zalex_photz.photoz < 1.5]
```

```python
zalex_photz['ratio_hbeta'] = zalex_photz.flux_hbeta.values / zalex_photz.flux.values
zalex_photz['ratio_hbeta'] = zalex_photz.ratio_hbeta / zalex_photz.ratio_hbeta.max()

zalex_photz['ratio_halpha'] = zalex_photz.flux_halpha / zalex_photz.flux
zalex_photz['ratio_halpha'] = zalex_photz.ratio_halpha / zalex_photz.ratio_halpha.max()

zalex_photz['ratio_oii'] = zalex_photz.flux_oii / zalex_photz.flux
zalex_photz['ratio_oii'] = zalex_photz.ratio_oii / zalex_photz.ratio_oii.max()

zalex_photz['ratio_oiii'] = zalex_photz.flux_oiii / zalex_photz.flux
zalex_photz['ratio_oiii'] = zalex_photz.ratio_oiii / zalex_photz.ratio_oiii.max()
```

```python
zalex_photz = zalex_photz.merge(df[['ref_id','bmu','x_cell','y_cell']], on ='ref_id' )
```

```python
Ld.shape
```

```python
ha_flux = zalex_photz.groupby(['x_cell','y_cell']).flux_halpha.median().reset_index()
ha_flux = ha_flux.drop_duplicates(['x_cell','y_cell'])
ha_flux = ha_flux.pivot(index = 'y_cell', columns = 'x_cell', values = 'flux_halpha')

L_ha = ha_flux.values * 4 * np.pi * Ld**2
L_ha = L_ha / np.nanmax(L_ha)

hb_flux = zalex_photz.groupby(['x_cell','y_cell']).flux_hbeta.median().reset_index()
hb_flux = hb_flux.drop_duplicates(['x_cell','y_cell'])
hb_flux = hb_flux.pivot(index = 'y_cell', columns = 'x_cell', values = 'flux_hbeta')

L_hb = hb_flux.values * 4 * np.pi * Ld**2
L_hb = L_hb / np.nanmax(L_hb)


oii_flux = zalex_photz.groupby(['x_cell','y_cell']).flux_oii.median().reset_index()
oii_flux = oii_flux.drop_duplicates(['x_cell','y_cell'])
oii_flux = oii_flux.pivot(index = 'y_cell', columns = 'x_cell', values = 'flux_oii')

L_oii = oii_flux.values * 4 * np.pi * Ld**2
L_oii = L_oii / np.nanmax(L_oii)

oiii_flux = zalex_photz.groupby(['x_cell','y_cell']).flux_oiii.median().reset_index()
oiii_flux = oiii_flux.drop_duplicates(['x_cell','y_cell'])
oiii_flux = oiii_flux.pivot(index = 'y_cell', columns = 'x_cell', values = 'flux_oiii')

L_oiii = oiii_flux.values * 4 * np.pi * Ld**2
L_oiii = L_oiii / np.nanmax(L_oiii)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (9,8))
ha = ax[0,0].imshow(L_ha.T, vmin = 0, vmax = .01)

hb = ax[0,1].imshow(L_hb.T, vmin = 0, vmax = .01)
oii = ax[1,0].imshow(L_oii.T, vmin = 0, vmax = .1)
oiii = ax[1,1].imshow(L_oiii.T, vmin = 0, vmax = .001)

ax[0,0].set_ylabel('y cell', fontsize = 14)
ax[1,0].set_ylabel('y cell', fontsize = 14)
ax[1,0].set_xlabel('x cell', fontsize = 14)
ax[1,1].set_xlabel('x cell', fontsize = 14)

ax[0,0].set_xticks([])
ax[0,1].set_xticks([])
ax[0,1].set_yticks([])
ax[1,1].set_yticks([])


plt.colorbar(ha, ax = ax[0,0], label=  'Luminosity [Normalised]')
plt.colorbar(hb, ax = ax[0,1], label=  'Luminosity  [Normalised]')
plt.colorbar(oii, ax = ax[1,0], label=  'Luminosity  [Normalised]')
plt.colorbar(oiii, ax = ax[1,1], label=  'Luminosity  [Normalised]')

ax[0,0].set_title(r'$H_{\alpha}$',fontsize = 14)
ax[0,1].set_title(r'$H_{\beta}$',fontsize = 14)
ax[1,0].set_title(r'$O_{ii}$',fontsize = 14)
ax[1,1].set_title(r'$O_{iii}$',fontsize = 14)
plt.savefig('/data/astro/scratch/lcabayol/Euclid/referee_report/figures/SOM_EL_cosmos.pdf', bbox_inches = 'tight')
```
